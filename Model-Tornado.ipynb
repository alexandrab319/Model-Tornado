# Model Tornado

## About

##### For this project, I will study a constructed model of the wind speed for a tornado. 
##### To model the movement of air in during a tornado, a vector field can be constructed to describe the flow of wind around its center. 
##### I was given a vector field describing the direction the wind is moving at a given point as well as a scalar function describing the wind-speed at each point. 
##### With these functions, I calculated the curl of the wind-field, the amount of work done on a lifted object, and the amount of wind flowing through three reference surfaces. 
##### I used these calculations to discuss the behavior of the weather event as predicted by the model.

## The Project

Let wdir(x, y, z) = ⟨−x − y, x − y, √2⟩ be a vector field representing the direction the wind is moving at a point (x, y, z). Is this vector field conservative? Find a unit-vector-field wdir.

#### The Field is not conservative, the plot shows the direction in which the wind moves

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x, y, z = np.meshgrid(np.arange(-3, 3, 1),
                      np.arange(-3, 3, 1),
                      np.arange(0, 3, 1))

u = (-1 * x - y) / ((-1 * x - y) ** 2 + (x - y) ** 2 + 2) ** 0.5
v = (x - y) / ((-1 * x - y) ** 2 + (x - y) ** 2 + 2) ** 0.5
w = 2 ** 0.5 / ((-1 * x - y) ** 2 + (x - y) ** 2 + 2) ** 0.5

ax.quiver(x, y, z, u, v, w, color='black')

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

The event horizon of the tornado is described by z = ln(x^2 + y^2) for x^2 + y^2 ≥ 1. Plot the event horizon.

#### Event horizon of the tornado

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

x, y = np.meshgrid(np.arange(-6, 6, 1), 
                    np.arange(-6, 6, 1))
                    
u = x
v = y
w = np.log(x**2 + y**2)

ax.plot_surface(u, v, w, color = 'black')

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

ax.view_init(60, 45)
plt.subplot();
```

 #### Amount of work the wind flow does on an object

To simplify the model, the force F on an object is proportional to the air flow w by a constant k related to the surface area of the object. In other words, we’ll say F(x, y, z) = kw. An object moves along a trajectory given by r(t) = r0 cos(t)bi + r0 sin(t)bj + tkb for 0 ≤ t ≤ 2π, where r0 is a constant. What is the amount of work W done by the air flow on this object?  Plot the trajectory for r0 = 1.

#### The movement and speed of the wind around the event horizon of the tornado

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

x, y = np.meshgrid(np.arange(-6, 6, 1), 
                    np.arange(-6, 6, 1))
                    
u = x
v = y
w = np.log(x**2 + y**2)

ax.plot_surface(u, v, w, color = 'royalblue', alpha = 0.5)

x, y, z = np.meshgrid(np.arange(-3, 3, 1), 
                      np.arange(-3, 3, 1), 
                      np.arange(0, 3, 1))

s = 10 
u = s * (-1 * x - y) / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5
v = s * (x - y) / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5
w = s * 2**0.5 / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5


for j in range(len(x)):
    for i in range(len(x[j])):
        for k in range(len(x[i,j])):
            if x[j][i][k]**2 + y[j][i][k]**2 >=1:
                ax.quiver(x[j][i][k], y[j][i][k], z[j][i][k], u[j][i][k], v[j][i][k], w[j][i][k], length=0.2, color='black')


plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")
ax.view_init(45, 45)
plt.subplot();
```

Plot the amount W done on the object as a function of k (related to its surface area) and r0 (the radius of its trajectory)

The trajectory of the object is instead given by r(t) = r0 sin(t)bi + r0 cos(t)bj + tkb for 0 ≤ t ≤ 2π. What is the amount of work W done on the object by the air flow. Plot the trajectory for r0 = 1

#### The plane illustrates the impact of the radius of the trajectory and the size of the force vector on the work performed on the object

```python
# (sqrt(2) + r^2) * r * e^-r * 2 * (pi)^2

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

k, r = np.meshgrid(np.arange(-3, 3, 0.2), np.arange(0, 3, 0.2))

u = k
v = r
w = k * ((2**0.5 + r**2) * r * np.e**(-r) * 2 * (np.pi)**2)

ax.plot_surface(u, v, w, color = 'black')

plt.xlabel("k")
plt.ylabel("$r_0$")
ax.set_zlabel("Z")

plt.subplot();
```

How does this differ qualitatively from the plot of work for the original trajectory?

```python
# (sqrt(2) + r^2) * r * e^-r * 2 * (pi)^2

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

k, r = np.meshgrid(np.arange(-3, 3, 0.2), np.arange(0, 3, 0.2))

u = k
v = r
w = k * ((2**1.5) * r * np.e**(-r) * 2 * (np.pi)**2)

ax.plot_surface(u, v, w, color = 'black', alpha = 0.8)

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.subplot();
```

Find the inward flux of air across the cylinder as a function of its radius r0. Plot this single-variable function for a range of r0 that makes sense. How does the flux qualitatively change as you move away from the center of the tornado?

#### The inward flux rallies as you move farther away from the tornado, quickly increasing between a radius of 2.5 and 5.0, before rapidly decreasing again

```python
fig = plt.figure(figsize = (10,10))

r = np.linspace(0, 20, 20)
y = np.pi**3 * r**3 * np.e**(-r)

plt.plot(r, y)
plt.xlabel("$r_0$")
plt.ylabel("Inward Flux")

plt.subplot();
```

Find the air flux upwards through the surface from the last part as a function of its height h. Plot this single-variable function. What

#### The upward air flux increases at a very low height, then decreases linearly after reaching a height of about 2
